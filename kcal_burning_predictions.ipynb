Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

Data - Height 6 Feet and Above
Use the chart below to estimate calories burned by step count if your height is 6 feet or more. This assumes that you take 2,000 steps per mile.

```python
steps = [1000,2000,3000,4000]

```

# 100lb

```python
df = {


    'steps':[5000,10000,15000,20000],

    'Burned_calories_by_wieght100lb':[138,275,413,550],
}

df100 = pd.DataFrame(df)

df100.to_csv('df100')

```

```python
x = df100[['steps']]

y = df100['Burned_calories_by_wieght100lb']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


model = LinearRegression()

model.fit(x_train, y_train)


y_pred = model.predict(x_test)




print (y_pred)

plt.scatter(x_test, y_test, color='blue')
plt.plot(x_test, y_pred, color='red')
```

```python
print(x_train)
```

the plot is behind it in the figure above

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
plt.scatter(y_test, y_pred, color='red')
plt.plot(y_test, color='blue')  # ideal line
plt.xlabel("Actual ")
plt.ylabel("Predicted ")
plt.title("Actual vs Predicted ")
plt.grid(True)
plt.show()
```

```python
new_x  = {

    'steps':[5000,4000]


}

new_y = pd.DataFrame(new_x)

n_x = new_y.loc[0:]

pred = model.predict(n_x)

```

For the 100lb the model predicted the actual data without seeing it bcs i passed an accurate mean values so it know the mean of the values even if its not passed in the training .
